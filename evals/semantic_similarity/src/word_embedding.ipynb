```python
import json

with open('../adjectives.json') as f:
    adjectives = json.load(f)['adjectives']

adjectives[:5]
```

```python
from sentence_transformers import SentenceTransformer
# model = SentenceTransformer("BAAI/bge-small-en-v1.5", trust_remote_code=True)
model = SentenceTransformer("nomic-ai/nomic-embed-text-v1", trust_remote_code=True)
```

```python
queries = adjectives

# system_prompt = "Please compare this word with others in the list"
# queries = [system_prompt + query for query in queries]

query_embeddings = model.encode(queries)
print(f'shape of query_embeddings: {query_embeddings.shape}')
```

```python
print(f"Similarity function: {model.similarity_fn_name}")

similarities = model.similarity(query_embeddings, query_embeddings)
print(similarities)
```
